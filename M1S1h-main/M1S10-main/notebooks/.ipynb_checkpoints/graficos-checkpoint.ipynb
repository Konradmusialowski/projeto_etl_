```python


import sys
import os
import pandas as pd
import matplotlib.pyplot as plt

from src import analysis,database
sys.path.append(os.path.abspath(os.path.join(os.getcwd(), '..')))

```

```python
data = analysis.movimentacao_por_centro()

df = pd.DataFrame(
    data,
    columns=[
        'centro',
        'regiao',
        'total_entrada',
        'total_saida'
    ]
)

df.plot(
    x='centro',
    y=['total_entrada', 'total_saida'],
    kind='bar',
    figsize=(10,5)
)

plt.title('Entradas e Saídas por Centro de Distribuição')
plt.ylabel('Quantidade')
plt.xlabel('Centro')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
